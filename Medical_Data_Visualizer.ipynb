```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("/content/medical_examination.csv")
```

```python
df.head(29)
```

```python
df.shape
```

```python
df = df.drop(df.columns[1], axis=1)
df
```

```python
df.info()
```

```python
df["weight"]=df["weight"].astype(int)
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr(),annot= True,fmt=".2f")
```

```python
sns.pairplot(df[["ap_hi","ap_lo","cholesterol","gluc","smoke","alco","active","cardio"]])
plt.show()
```

```python
df["cholesterol"].plot(kind="density",color="pink",alpha=1,figsize=(10,5))
```

```python
df['gluc'].value_counts().plot(kind="barh",color="red",alpha=1,figsize=(10,5))
```

```python
df["smoke"].value_counts().plot(kind="barh", color="skyblue", alpha=1, figsize=(10, 5))
```

```python
df['cardio'].value_counts().plot(kind="pie")
```

```python
plt.violinplot(df[["ap_hi","ap_lo","cholesterol","gluc"]])
```

```python
sns.distplot(df["ap_hi"])
```

```python
sns.jointplot(x="smoke",y="cardio",data=df,kind="kde")
```

```python
sns.catplot(x="cholesterol",y="cardio",data=df,kind="bar")
```

```python
sns.jointplot(x="ap_hi", y="cardio", data=df, kind="reg", height=8)
plt.show()
```

```python
sns.stripplot(x="alco", y="cardio", data=df)
```

```python
g = sns.FacetGrid(df, col="alco", col_wrap=4)
g.map(sns.barplot, "cardio", "cholesterol")
```

```python
sns.ecdfplot(data=df[["ap_hi","ap_lo","alco","cardio","smoke"]])
```
