```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("/content/data_science_salaries.csv")
df.head(13)
```

```python
df.shape
```

```python
df.info()
```

```python
group=df.groupby("job_title")["salary_in_usd"].mean()
group
```

```python
group.describe().T
```

```python
plt.figure(figsize=(45,13))
group.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Average Salary by Job Title", fontsize=14)
plt.xlabel("Job Title", fontsize=12)
plt.ylabel("Average Salary (USD)", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(45,13))
df["job_title"].value_counts().plot(kind="bar", color="orange", edgecolor="black")
plt.title("Job Title Distribution", fontsize=14)
plt.xlabel("Job Title", fontsize=12)
plt.ylabel("Count", fontsize=12)
```

```python
df["company_location"].value_counts().plot(
    kind="pie",
    autopct='%1.1f%%',
    figsize=(6, 6),
    startangle=90,
    colors=["yellow", "lightgreen", "violet","pink","purple","indigo","lavender","gold","silver","green"],
    title="Distribution of Companies by Location"
)
plt.tight_layout()
plt.show()
```

```python
colors=["red","yellow","green"]
df["company_size"].value_counts().plot(kind="barh",color=colors,edgecolor="black")
plt.title("Company Size Distribution")
plt.xlabel("Count")
plt.ylabel("Company Size")
```

```python
df["experience_level"].value_counts().plot(kind="barh",color=["purple","indigo","green","red"],edgecolor="black",alpha=1)
plt.title("Experience Level Distribution")
plt.xlabel("Experience Level")
plt.ylabel("Count")

```

```python
sns.violinplot(x=df["employment_type"], palette="muted", inner="point", edgecolor="black")
plt.title("Employment Type Distribution")
plt.xlabel("Employment Type")
plt.ylabel("Count")
plt.show()
```

```python
df["work_models"].value_counts().plot(kind="barh",color=["orange","skyblue","green"],edgecolor="black",alpha=1)
plt.title("Work Model Distribution")
plt.xlabel("Work Model")
plt.ylabel("Count")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x=df["experience_level"],y=df["salary_in_usd"],color=["lightpink"],alpha=1)
plt.title("Experience Level vs Salary")
plt.xlabel("Experience Level")
plt.ylabel("Salary (USD)")
```

```python
sns.pairplot(df,hue="experience_level")
```

```python
fig= plt.subplots(figsize=(20,30))
sns.scatterplot(x="salary_in_usd",y="company_location",hue="experience_level",data=df)
plt.title("Salary vs Company Location")
plt.xlabel("Salary (USD)")
plt.ylabel("Company Location")
plt.tight_layout()
plt.show()
```

```python
df.hist()
plt.tight_layout()
plt.show()
```

```python
sns.swarmplot(x="experience_level",y="salary_in_usd",data=df)
plt.title("Salary vs Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Salary (USD)")
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(x="work_models",y="salary_in_usd",data=df)
plt.title("Salary vs Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Salary (USD)")
plt.tight_layout()
plt.show
```
